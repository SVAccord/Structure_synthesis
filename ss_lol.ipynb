```python
''' try to get all realized structures from eqg table on the field of eq. groups
In this program do throught the list of lists'''

import numpy as np
import pandas as pd
import string
import time
#import math
#import re

def listToString(list1):
    return (''.join(map(str, list1)))

def listToString_s(list1):
    return (' '.join(map(str, list1)))

num2alpha = dict(zip(range(1, 27), string.ascii_lowercase))

def digit2alpha(txt):
    '''converts string of digits into letters'''
    # add defence from input 0 and bigger than 26
    my_list = txt.split()
    txt_res = ''
#     print(my_list)
#     print(num2alpha[int(my_list[0])])
    for i in range(len(my_list)):
        txt_res += num2alpha[int(my_list[i])] + ' '
    #txt = listToString_s(my_list)
    return txt_res.rstrip()

def checkinpattern_str(txt, pattern):
    # get strings
    # if work w list from begining will be faster
    ltxt = txt.split()
    stxt = set(ltxt)
    return stxt.issubset(set(pattern))

def checkinpattern_list(ltxt, lpattern):
    # get lists
    stxt = set(ltxt)
    return stxt.issubset(set(lpattern))

print(checkinpattern_str('31 22', '1 2 3'))
print(checkinpattern_str('3 2', '1 2 3'))
print(checkinpattern_list([1, 2], [1, 2, 3]))

def style_ml_pair(text):
    # проще через сплит
    txt = 'M'
    i = 0
    while i < len(text):
        if text[i].isdigit() :
            txt += 'l' + text[i]
            i += 1
            continue
        else: 
            i += 1
    return txt

def style_ml_r(note, reminder):
    txt = 'Ml' + listToString(reminder) + note
    return txt

def style_ml(note, reminder):
    txt = 'M' + note + reminder
    return txt

#result = re.search('1 3', '1 2 3')
#if result:
#    print(result[0])

style_ml_pair('1 2')

def getnumoffirstleave(text):
    '''input format like: M1l1l2...'''
    i = 0
    num = 0
    while i < len(text):
        if text[i] == 'l':
            if i+1 <= len(text):
                if text[i+1].isdigit() :
                    if i+2 <= len(text):
                        j = i + 2
                        while text[j].isdigit():
                            j += 1
                            if j == len(text):
                                break
                        num = int(text[i+1:j])
                        break
                    else:
                        num = int(text[i+1])
                        break
        i += 1
    return num

def namem(text):
    txt = ''
    cnt = 1
    for i in range(len(text)):
        if text[i] == 'M':
            txt += 'M' + str(cnt)
            cnt += 1
        else:
            txt += text[i]
    return txt
```

```python
eqglist = pd.read_excel('210827_psy_ag.xlsx')
#eqglist = pd.read_excel('210706_psy_allgroups_eqg_res_s.xlsx')
#eqglist = pd.read_excel('210805_postcode35_l6_ag.xlsx')
#eqglist = pd.read_excel('210702_arct_allgroups_eqg_res_min.xlsx')
lnum = 10
scale = 4
# better than set lnum - to find it from the data

eqglist = eqglist[eqglist.iloc[:,1] <= scale]
eqglist.reset_index(inplace = True, drop = True) 
# sort groups not in scale
eqglist = eqglist.iloc[:,0]

if len(eqglist.iloc[len(eqglist)-1].split()) < lnum-1:
    print("no available structures in input data")
else:
    # add full list structure on the last string
    eqglist[len(eqglist)] = listToString_s([i for i in range(1, lnum+1)])

print("len of data in scale:", len(eqglist))
print("last string:", eqglist.iloc[len(eqglist)-1])
#print(eqglist)
#convert string of digits into letters
# for i in range(0,len(eqglist)):
#     eqglist[i] = digit2alpha(eqglist[i])
# print(eqglist)

# convert eqglist to list
for i in range(0,len(eqglist)):
    eqglist[i] = eqglist[i].split()
print(eqglist)

lol = [[] for i in range(len(eqglist))]

a = 0 
while len(eqglist[a]) <= 2:
    lol[a].append(style_ml_pair(eqglist[a]))
    a += 1
    
print(lol)

# create list of used groups
used_groups = [[] for i in range(len(eqglist))]
```

```python
# теперь заполняем тройки и выше
start = time.time()
for i in range(a, len(eqglist)):
    pattern = eqglist[i]
    #print('p', pattern, " ----------------------------")
    #print("i", lol[i])
    j = 0
    while len(eqglist[j]) < len(pattern):
        #print(' seek', eqglist[j])
        #print(" ", lol[j])
        # просматриваем записи с начала таблицы
        if checkinpattern_list(eqglist[j], pattern):
            reminder  = set(pattern) - set(eqglist[j])
            #print('  rem', reminder)
            #if len(pattern) - len(eqglist[j]) >= 2:
            if len(reminder) >= 2:
                # пробуем найти составное дополнение k eqglist[j]
                n = j + 1 # похоже что это условие убирает необходимсть отмечать использованные дополнения
                # но, проверять в симметричных парах очередность все-равно нужно
                # кажется, что начинать можно не с +1 а сначала групы размером = остатку
                while len(eqglist[n]) < len(pattern):
                    #if checkinpattern_list(eqglist[n], reminder): # не просто содержится в остатке, а равна ему
                    if set(eqglist[n]) == reminder:
                        # used_groups[n] = 1
                        # из двух ветвей собать запись
                        for jj in range(len(lol[j])):
                            for nn in range(len(lol[n])):
#                                 print('   j', j)
#                                 print('   n', n)
#                                 print('   l1', lol[j])
#                                 print('   l2', lol[n])
                                # нужна проверка для номеров симметричных ветвей - сперва меньший
                                if len(lol[j][jj]) == len(lol[n][nn]):
#                                     print(lol[j][jj])
#                                     print()
                                    if getnumoffirstleave(lol[j][jj]) < getnumoffirstleave(lol[n][nn]):
                                        lol[i].append('M' + lol[j][jj] + lol[n][nn])
                                    else:
                                        lol[i].append('M' + lol[n][nn] + lol[j][jj])
                                else:
                                    lol[i].append('M' + lol[j][jj] + lol[n][nn])
                                #lol[i].append('M' + lol[j][jj] + lol[n][nn])
                                #print('   2b', lol[i])
                    n += 1
            else:
                # из записи и листа собрать запись
                for o in range(len(lol[j])):
                    lol[i].append(style_ml_r(lol[j][o], reminder))
                    #print('   +r', lol[i])
        j += 1
    #print("i", lol[i])

#lol
finish = time.time()
result = finish - start
print("Program time: " + str(result) + " seconds.")
```
